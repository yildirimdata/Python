**Escape Sequences**

- \' (ters slash ve ')  This represents a single quote
- \n   This represents a new line
- \r   This represents a carriage return (satƒ±r ba≈üƒ±na getirme)
- \t   This represents a tab - four spaces
- \b   This represents a backspace
- \f   This represents a formfeed (sayfa ilerletme, sayfa ba≈üƒ± tu≈üu) 
- \ooo This represents an octal value (sekizli-oktal)
- \xhh This represents a hex value (hex v: onaltƒ±lƒ±k deƒüer)
- \\   This represents a backslash
- \uxxxx  This represents 16 bits hex value
- \uxxxxxxxx  This represents 32 bits hex value

**Arithmetic Operations**

```python
# Any mathematics operator can be used before the = character to make an in-place operation
# -= decrements the variable in place, (decrement: azaltma, eksiltme - decrease)
# += increment the variable in place, (increment: artƒ±rma, toplama - increase)
# *= multiply the variable in place,
# /= divide the variable in place,
# //= Ô¨Çoor divide the variable in place,
# %= returns the modulus of the variable in place,
# **= raise to power in place.

x = 5
print('value of x     : ', x)
x += 2
print('2 more of x    : ', x, "\n")  # using string expression '\n', we produce an extra empty line
x -= 3
print('3 minus x      : ', x, "\n") 
y = 10
y *= 2
print('2 times y      : ', y, "\n")  # yukarƒ±dakilerin hepsiyle yapƒ±labilir.
```

**Boolean Operations**

```python
# Python has three built-in boolean operators: 
# and, or and not. 
# Except not, all are binary operators, which means two arguments are required.
# And operator : The and operator evaluates all expressions and returns the last expression if all expressions are evaluated True. 
# Otherwise, it returns the Ô¨Årst value that evaluated False.
# Or operator : The or operator evaluates the expressions left to right and 
# returns the Ô¨Årst value that evaluated True or the last value (if none is True).
# not operator :it evaluates the expression that follows it as the opposite of the truth. eg: not True means False
```
