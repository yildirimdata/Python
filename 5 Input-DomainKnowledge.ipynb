Q: Write a program that multiplies three numbers entered by the user. Print the output with the format method.

```python
# Sorunun cevabı. ilk önce str'yi dikkate almayan yanlis cevap:

num1 = input("enter num1: ")
num2 = input("enter num2: ")
num3 = input("enter num3: ")  # bu ücünü yazip calistirnca asagida sirayla 3üne deger atamamizi isteyecek
```

```python
# şimdi burada hata çıkacak
output = num1 * num2 * num3
# hata tipi: TypeError: can't multiply sequence by non-int of type 'str'
# bu 3-4-5 str. cunku input hep str verir. Dolayısıyla once int'e cevrilmeli. Below:
```

```python
output = int(num1) * int(num2) * int(num3)  # her birini inte cevirince outputa cevap degeri eklenir
# outputu print ederek cevabı buluruz
print(output)
```

```python
# sonucu da su sekilde yazıya dökeriz:

print("the multiplication of num1, num2 and num 3 is equal to : ", output)
```

```python
# ikinci bir yöntem de önce hepsini tek tek en basta int'e cevirmek
num1 = int(input("enter num1: "))
num2 = int(input("enter num2: "))
num3 = int(input("enter num3: "))

# aşağıda degerleri yaz kutusu cikacak ama artik onlar int. hemen altta carpalim mesela direkt multiplies
# dolayısıyla burada output variable atamaya da gerek kalmamis olur
```

Problem 2

Write a program that calculates body mass index from height and weight entered by the user.

(soruyu çözmeden önce domain knowledge yaparız)

**Domain Knowledge**

```python
# alan bilgisi demek. o alanla ilgili bilgi.
# task alınca once alanla ilgili bilgi toplamalı, daha sonra kalem kagitla kafada
# problemin cozumu tartisilmali. bir algoritma, neler yapilacagi taslagi hazirlanir.
# ondan sonra kod yazılır.
# bir kere kullanacagimiz sey icin variable atamaktansa direkt print fonksiyonu kullanabilir. 
# hafızada her degisken bir yer kaplar. BMI orneginde body_mass_ıdx atamadan yapılan cozume bak, daha iyi o
```

Yaptığım alan araştırması neticesinde vücut kütle indeksinin formülünün şu şekilde olduğunu buldum:
    
Body mass index :  Weight / Height(m) * Height(m)

```python
weight = float(input("enter your weight (for example 75) : "))
height = float(input("enter your height (for example 1.85): "))
# sonra bmi variable atayip formulumuzu yazarız.
# direkt floata neden cevirdik: cunku str verecek. str carpmaz bolmez. bir sonraki adimda yapmaktansa simdiden yaptik.
body_mass_idx = weight / (height ** 2)
# değişken ismini belirlerken kurallara uydum. 
# değişken ismi içeriği ile ilgili fikir veriyor.,
# python'un keywordlerinden biri değil.
# rakamla başlamıyor.
# içinde özel karakter yok.
# çalıştırınca input oldugu icin ilk asagida deger isteyecek 

```

```python
print("Your body mass index is ", body_mass_idx)  # sonra istersek body-mass-idx variableini printle yazdırırız
```

```python
print("Your body mass index is ", weight / (height ** 2))
# ya da istersek direkt print fonksiyonunun içinde bu hesaplamayı yaparız
# böyle yaparsak böylece bosuna body-mass-ıdx diye bir variable atamamamis oluruz
# her gereksiz variable bir memory reserve eder. milyon datali bir projede bu cok alan demektir
```

**Problem 3**

```python
# burada hemen aklımıza floor division ve modulus gelir
print("You can buy", 200 // 11, "pieces with your money." )
print ("After buying 18 pieces,", 200 % 11, "Dollar you have left.")
```

**Problem 5**

Get word, separator and repetition values from the user and print the word on the screen according to the given values.

```python
word = input("enter your text here for word ")  # input zaten str verecek degistirmeye gerek yok
separator = input("enter your text here for separator ")  # bu da str olmalı
repetition = int(input("enter repetition number here "))  # repetition no will be a number, so it must be converted to int

print((word + separator), * (repetition - 1), + word)  # rep -1 yaptik cunku yoksa aralarda top 4 olacak. en son :) ile bitmesin
```
