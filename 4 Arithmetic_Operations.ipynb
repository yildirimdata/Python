```python
degisken = 10
print(degisken)
```

```python
degisken = degisken + 1  # esitligin sag tarafi value. sol da variable.
print(degisken)
# her print yaptigimda degiskeni bir artiriyor. ilk tiklama 11 dedi sonra 12 sonra 13 sonra 14 
```

```python
degisken = degisken - 1
print(degisken)  # simdi 1 eksilttim, tek tikladim 13 oldu.. yeni degisken artik 13, bir daha print dersek 12, bir daha 12....
```

```python
# bunu su sekilde de yapabiliriz. + isareti equation oncesine gelir
degisken = 10
degisken += 1
print(degisken)
```

```python
a, b = 8, 2
a += b  # a ile b degiskenini topla demek oluyor bu. yani yeni a a'nın b (2) fazlası oluyor
# 2 yazmamak lazim. cunku b asagilarda degisebilir ve a da ona gore degissin diye b yazarız
# a += b demek a = a+b demek
print(a)
```
