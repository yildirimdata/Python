Indexing and slicing strings

** string indexing : square brackets [ ] ile yapÄ±lÄ±r

```python
name = 'Yildirim'

print("His name is :", name)
print("Second letter of his name is :", name[1])
print("First letter of his name is :", name[0])
print("Last letter of his name is :", name[-1])
print("the letters from 3rd to 5th in his name is:", name[2:4])
print("the last 2 letters of his name is:", name[-2:])
print("his full name is:", name[:])
```

The formula **syntax of string indexing is : string[start:stop:step].**

indexlemede hep saÄŸa doÄŸru gider. sadece step eksi olursa terse gider. Ornegin ::-1 en tersten baÅŸlayÄ±p baÅŸa kadar yazar.


Bu Ã¼Ã§lÃ¼de start yerine yazdÄ±ÄŸÄ±n nerden baÅŸlayacaÄŸÄ±nÄ±, stop yerine yazdÄ±ÄŸÄ± nerde duracaÄŸÄ±nÄ±, step de hangi adÄ±m aralÄ±klarÄ±yla gideceÄŸini sÃ¶yler. Ã–rneÄŸin 2:5:2 ise 2 den start 5 te stop ama atlayarak demek(bu Ã¶rnekte 2den 4e). step 1 yazÄ±lmaz cÃ¼nkÃ¼ o zaten default, atlama sÄ±rayla git demek. ::2 demek baÅŸtan baÅŸla sona kadar git ama 2 ÅŸer aralÄ±k.

string[:] : returns the full copy of the sequence

string[start:] : returns elements from start to the end element

string[:stop] : returns element from the 1st element to stop-1

string[::step] : returns each element with a given step (stride: atlama, uzun adÄ±mlarla yÃ¼rÃ¼me)

```python
team = "Galatasaray"
# G0 a1 l2 a3 t4 a5 s6 a7 r8 a9 y10
# y-1 a-2 r-3 a-4 s-5 a-6 t-7 a-8 l-9 a-10 G-11
# negative enumeration starts from -1, thus the number of the last letter will be one bigger than + enum.  
print("Name of the winning team is", team[:]) # gives the whole
print(team[1:])  # numaralandÄ±rma sÄ±fÄ±rdan baÅŸladÄ±ÄŸÄ± iÃ§in 1 ikinci harf olur # starts from index 1 to the end
print(team[:6])  # aslÄ±nda durduÄŸu 6 s oluyor. dur dedigini gostermiyor. # starts from zero to 5th index
print(team[::2]) #starts from zero to end by two step
print(team[1::2])  # starts from index 1 to the end by 2 step
print(team[-3:])  # starts from index -3 to the end
print(team[::-1])  # negative step starts from the end to zero: tersine cevirir

name= "Kadir Yildirim"
print(name[:-1])  # it stops at -1
print(name[::-1])  # negative step starts from the end to zero: tersine cevirir
```

**String Formatting with '%' Operator**

ğŸ‘‰ğŸ»% operator gets the values in order and prints them in order using several characters accordingly.

Eskiden kullanÄ±lÄ±yordu. Ama bilelim. 

```python
phrase = 'I have %d %s %d %s until the end of the %s' % (2, "hours", 35, "minutes", "lesson")
print(phrase)
```

```python
# We can limit the character numbers of the strings by using % operator
statement = "Galatasaray is the winner."  # we get first 11 characters of the string
print("%.11s" % statement)  # yukarÄ±daki %.2f ile aynÄ± mantÄ±k. burada str'i ilk 11 ile limit ediyoruz.
# ama noktaya dikkat. print("%11s" % statement) yazdÄ±m ve "Galatasaray is the winner." tamamÄ±nÄ± return etti.
```

```python
# We can also use variables with ğŸ‘‰ğŸ»% operator to format the string.
# Tipp: we assign values to variables in curly braces '{}'. This format is a dictionary type

print('%(amount)d pounds of %(fruit)s left' % {'amount': 33, 'fruit':'bananas'})
print('The champion of the %(season)s is %(team)s.' % {'season' : '2020-2021', 'team' : 'Galatasaray'})
```
