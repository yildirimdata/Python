Indexing and slicing strings

** string indexing : square brackets [ ] ile yapılır

```python
name = 'Yildirim'

print("His name is :", name)
print("Second letter of his name is :", name[1])
print("First letter of his name is :", name[0])
print("Last letter of his name is :", name[-1])
print("the letters from 3rd to 5th in his name is:", name[2:4])
print("the last 2 letters of his name is:", name[-2:])
print("his full name is:", name[:])
```

The formula **syntax of string indexing is : string[start:stop:step].**

indexlemede hep sağa doğru gider. sadece step eksi olursa terse gider. Ornegin ::-1 en tersten başlayıp başa kadar yazar.


Bu üçlüde start yerine yazdığın nerden başlayacağını, stop yerine yazdığı nerde duracağını, step de hangi adım aralıklarıyla gideceğini söyler. Örneğin 2:5:2 ise 2 den start 5 te stop ama atlayarak demek(bu örnekte 2den 4e). step 1 yazılmaz cünkü o zaten default, atlama sırayla git demek. ::2 demek baştan başla sona kadar git ama 2 şer aralık.

string[:] : returns the full copy of the sequence

string[start:] : returns elements from start to the end element

string[:stop] : returns element from the 1st element to stop-1

string[::step] : returns each element with a given step (stride: atlama, uzun adımlarla yürüme)

```python
team = "Galatasaray"
# G0 a1 l2 a3 t4 a5 s6 a7 r8 a9 y10
# y-1 a-2 r-3 a-4 s-5 a-6 t-7 a-8 l-9 a-10 G-11
# negative enumeration starts from -1, thus the number of the last letter will be one bigger than + enum.  
print("Name of the winning team is", team[:]) # gives the whole
print(team[1:])  # numaralandırma sıfırdan başladığı için 1 ikinci harf olur # starts from index 1 to the end
print(team[:6])  # aslında durduğu 6 s oluyor. dur dedigini gostermiyor. # starts from zero to 5th index
print(team[::2]) #starts from zero to end by two step
print(team[1::2])  # starts from index 1 to the end by 2 step
print(team[-3:])  # starts from index -3 to the end
print(team[::-1])  # negative step starts from the end to zero: tersine cevirir

name= "Kadir Yildirim"
print(name[:-1])  # it stops at -1
print(name[::-1])  # negative step starts from the end to zero: tersine cevirir
```

**String Formatting**

There are several ways in Python that we use when processing and using string data structures. The most important of these are:

2. % operator formatting,

3. string.format() method,

4. f-string formatting.



**Arithmetic Syntax**

```python
# +, -, *
a = "Galata"
b = "saray"
print(a+b)  # We can use + operator for combining the two string together without any spaces. to concate 
print(3* (a+b))  # We can also use * operator for repeating the string without any spaces.
```

```python
# As with numeric types, we can do addition operation in-place either with string type using 👉🏻+=
# a = a + b ile a += b demek aynı şey


fruit = "Apple"
print(fruit)
fruit += " Banana"
print(fruit)
fruit += " Orange"
print(fruit)
fruit += " Grape"
print(fruit)
fruit += " Melon"
print(fruit)

```

```python
str_one = 'upper'
print(* str_one)  # yalnız bir * operatoru valuenin elementlerine ayırır. aradakiler space characteri degil. len yine aynı yani
# * operatoru iterablei elementlerine ayırır.
aaa = "I'm angry"
print(* aaa)
# burada m ve a arası space var, return ederken orayı 3 spaces return eder.
print(* "I'm angry")  # sadece variable degil, valueyu yazarsak yine return eder.
```

**String Formatting with '%' Operator**

👉🏻% operator gets the values in order and prints them in order using several characters accordingly.

Eskiden kullanılıyordu. Ama bilelim. 

```python
phrase = 'I have %d %s %d %s until the end of the %s' % (2, "hours", 35, "minutes", "lesson")
print(phrase)
```

```python
phrase = 'I have %d %s and %.2f brothers' % (4, "children", 5) 
print(phrase)
# In the '%s' syntax : s stands for 'string'.
# In the '%.2f' syntax : f stands for 'float'. In this example 2 digits after point.
# %f yazdım 5.000000 brothers return etti. %.1f yazdım 5.0 return etti.
# In the '%d' syntax : d stands for 'numeric'.
```

```python
# We can limit the character numbers of the strings by using % operator
statement = "Galatasaray is the winner."  # we get first 11 characters of the string
print("%.11s" % statement)  # yukarıdaki %.2f ile aynı mantık. burada str'i ilk 11 ile limit ediyoruz.
# ama noktaya dikkat. print("%11s" % statement) yazdım ve "Galatasaray is the winner." tamamını return etti.
```

```python
# We can also use variables with 👉🏻% operator to format the string.
# Tipp: we assign values to variables in curly braces '{}'. This format is a dictionary type

print('%(amount)d pounds of %(fruit)s left' % {'amount': 33, 'fruit':'bananas'})
print('The champion of the %(season)s is %(team)s.' % {'season' : '2020-2021', 'team' : 'Galatasaray'})
```

**String Formatting with 'f-string'**

f-string is the string syntax that is enclosed in quotes with a letter f at the beginning. Curly braces 👉🏻{} that contain variable names or expressions are used to replace with their values.

```python
fruit = 'Orange'
vegetable = 'Tomato'
amount = 6
output = f"The amount of {fruit} and {vegetable} we bought are totally {amount} pounds"

print(output)
```

```python
print(f"2**3")  # bu 8 olmaz. curly braces yok. string olarak cikti verir.
print(f"{2**3}")  # curly braces olunca bu 8 olur
```

```python
result = f"{4*5}" # f'ten sonrasını quotes içine almayı unutma. error verir yoksa.
print(result)

team = 'GALATASARAY'
print(f"The champion is {team.capitalize()}")  # variable ismi sonrası capitalize() yazınca sonuc: 
# capitalize sonunda () olmazsa error.
print(f"The champion is {team.lower()}")  # variable ismi sonrası lower() yazınca sonuc: 
# variable.upper() da büyük yapar
```

```python
# We can also write multiline f-string. there are two ways for that: parantheses and backslash \
name = "Joseph"
job = "teachers"
domain = "Data Science"
message = (
     f"Hi {name}. "
     f"You are one of the {job} "
     f"in the {domain} section."
)
print(message)
```

```python
team = "Galatasaray"
league = "Turkish league"
season = "2020-2021 season"
country = "Turkey"
result = (
    f" {team}"
    f" is the champion"
    f" of the {league}"
    f" in {country}"
    f" in the {season}."  # her yeni satır başında veya üsttekinin sonunda boşluk bırakmayı unutma. bitişik çıkar yoksa.
)
print(result)
```

```python
# 2nd way: We can also use backslash 👉🏻\ between lines instead of parantheses for the multiple f-string formating lines.
name = "Joseph"
job = "teachers"
domain = "Data Science"
message = f"Hi {name}. " \
     f"You are one of the {job} " \
     f"in the {domain} section."

print(message)
```
