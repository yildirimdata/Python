#### <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">1. "myscript" adında bir modül oluşturalım</p>

```python
"""Bir modül olarak kullanılmak üzere dizayn edilen bir Script geliştiriyoruz..

Dikkat! 
Bir dosya; onu modül olarak import etmek istediğinizde,
ilk içe aktarma (import) için bir script (komut dosyası) olarak davranır.
Böylece scriptimiz içindeki print yaptıran kodlar "modül" ilk kez import edildiğinde yürütülür.
Oysa bir modülün import edildiğinde çıktı üretmesini istemeyiz.
```

#### <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">4. if __name__ == "__main__": kullanımını anlamak için şu örneği inceleyelim:</p>

````python
# Python program to execute main directly

print ("Bu her halükarda çalışır.")
 
if __name__ == "__main__":
    print ("Bu satır script doğrudan çalıştırıldığında yürütülür.\
Çünkü o zaman name == main'dir.")
else:
    print ("Bu satır yalnızca dosya import edilmişse çalışır.")
````

**Dikkat ederseniz bu script’i herhangi bir modül import etmeden çalıştırdığımız için; import edilmesi halinde yürütülen kod bloğu çalışmadı.**

#### <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">5. Oluşturduğumuz modüldeki my_function üzerinden if __name__ == "__main__": örneği yapalım: </p>

- Şimdi modülümüzün içinde tanımladığımız fonksiyonu (my_funcytion);
    - Önce **(modülü import etmeden) direkt Script’imizin içinde çalıştıracağız.** 
    - Sonra da **myscript.py modülünü import ederek çalıştıracağız.**


- Fonksiyonu daha önce bu notebook’ta çalıştırmadığımız için bu aşamada **"Name Error" exception'ı yükselecektir.** Bu nedenle;
    - Exception'ın Programı sonlandırmaması için **try - except** bloğu kullanarak (**exception handling yaparak**) sonraki satırdan itiraben programın devam etmesini sağlayacağız. 
    - Devamında modülümüzü import ederek içindeki fonksiyonun çalışmasını sağlayacağız.


````python
# Now if we want to use that module by importing:

try:
    if __name__ == "__main__":
        my_function()
except NameError:
    print("__name__, şuan __main__'de olduğu için \
my_function() bağımsız olarak yürütülmeye çalışıldı! \
Fakat Script içinde my_function tanımlı olmadığından Name Error yükseldi!\
Program sonraki satırdan devam ediyor..\n")

finally :
    print("""Bir Python kuralı olarak, .py uzantılı oluşturduğunuz dosya\ 
bir modül olarak içe aktarıldığında; Python, __name__ değişkenini 
modülün adına ayarlar. Ancak, dosya bir komut dosyası olarak çalıştırılırsa,
__name__ değişkeni '__main__' string değerine ayarlanır.\n""")
    
import myscript  # wordpad'de oluşturduğumuz modülü import etttik.
print("Alt satırda pyscript modülü içindeki fonksiyon çalıştırılacak..")
myscript.my_function()
````

**Dikkat! Outputun son satırındaki modül isminin (myscript), my_function() fonksiyonunun içinde __name__ attribute'ı ile yazdırıldığını hatırlayalım.**
