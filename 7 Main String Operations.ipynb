**Main String Operations**

String processing is of great importance in the world of Data Science as a field of research and business in the world. 

A significant thing to keep in mind is that string is an immutable data type. This means you can't just change the string in place, so most string methods return a copy of the string. Well, how can we do that?

This is the way you should follow: You must create a new variable for the copy you made or assign the same name to the copy to save the changes made to the string for later use.

**Searching a String**

To search patterns in a string there are two useful methods called startswith() and endswith() that search for the particular pattern in the immediate beginning or end of a string and return True if the expression is found.

The formula syntaxes are :

string.startswith(prefix[, start[, end]])

string.endswith(suffix[, start[, end]])

```python
state = "Bugün havanın sıcak olduğu günlerden bir ekim günü."
print(state.endswith('kim'))  # state variable kim ile bitmedigi için False verdi
print(state.endswith('ünü.'))  
print(state.startswith('Bug'))
```

These methods have optional arguments start and end. We can specify the search by adding arguments so that the area of search is delimited by start and end arguments.

Remember! Characters of string count from left to right and start with zero.

```python
state = "Bugün havanın sıcak olduğu günlerden bir ekim günü."
print(state.startswith("sıc", 14))  # index no 14 is s. If we said ("sıc", 15) it'd return F

# Mesela 5 ve 10 nolu indexler arasına dair bir şey soralım
print(state.endswith("a", 5, 10))  # a ile bittiği içn T verir
print(state.endswith("o", 5, 10))
```

**isnumeric() method**:

The isnumeric() method **returns True if all the characters are numeric** (0-9), otherwise False. Exponents, like ² and ¾ are also considered to be numeric values. **Its syntax is: string.isnumeric()**

"-1" and "1.5" are NOT considered numeric values, because all the characters in the string must be numeric, and the - and the . are not.

```python
txt = "565543"
x = txt.isnumeric()
print(x)
c = "10km2"
d = "-1"
e = "1.5"
print()
print(c.isnumeric())
print(d.isnumeric())
print(e.isnumeric())
```

**Changing a String**

The methods described below return the **copy** of the string with some changes made:

The formula syntax is : string.method()

some common and the most important **methods** of string changing :

**str.replace(old, new [,count])** replaces all occurrences of old with the new. The count argument is optional, and if the optional argument count is given, only the first count occurrences are replaced. count: Maximum number of occurrences to replace. -1 (the default value) means replace all occurrences.

**str.swapcase()** converts upper case to lower case and vice versa.

**str.capitalize()** changes the first character of the string to the upper case and the rest to the lower case.

**str.upper()** converts all characters of the string to the upper case.

**str.lower()** converts all characters of the string to the lower case.

**str.title()** converts the first character of each word to upper case.

IMP: note that, source text is unchanged. They are all copies of the source text.

```python
source = "Today I will learn some new issues about Python."
print(source)
print(source.replace("learn", "teach" [:]))  # ilk denemede quotes icinde yazmadim learn ve teachi, 
# bu nedenle learn is not defined dedi variable sanıp. yani variable tanıtmış olsak onla da yapabiliriz
# en sondaki [ ] vermezsen default olarak en sona kadar bakar.
print(source.swapcase())  # küçüğü büyüğe büyüğü küçüğe
print(source.capitalize())  # ilk element büyük gerisini küçük
print(source.upper())  # tamamı büyük
print(source.lower())  # tamamı küçük
print(source.title())  # her kelimenin ilk karakterini upper yapar.
print(source.find("will"))  # index nosunu verir
print(source.count("new"))  # stringde kaç tane new olduğunu sayar
print(* source)
print(source)  # yine orjinali unchanged olarak yazdırır unutma. yukardakiler hep copy.
```

**Editing a String**

The methods described below remove the trailing characters (i.e. characters from the right side). The default for the argument chars (chars: characters) is also whitespace. If the argument chars aren’t specified, trailing whitespaces are removed.

The formula syntax is : string.method()

str.strip() : removes all spaces (or specified characters) from both sides.

str.rstrip() : removes spaces (or specified characters) from the right side.

str.lstrip() : removes spaces (or specified characters) from the left side.

```python
source = "   Today I will learn some new issues about Python.   "
print(source.strip())  # removes all spaces from both sides. parantez icine başka birşey tanımlarsan onu remove eder.
print()
source = "Today I will learn some new issues about Python."
print(source.strip("about Python."))  # en sondan bu iki kelimeyi kaldırdı
print(source.strip("about Python. Today"))  # son iki ve takip eden baştaki ilk kelimeyi kaldırdı.
print(source.rstrip("."))
print(source.lstrip("To"))
```

```python
source_string = "interoperability"
print(source_string.strip("yi"))  # removes trailing "y" or "i" or "yi" or "iy" from both sides
print(source_string.lstrip("in"))  # removes "i" or "n" or "in" or "ni" from the left side
print()
space_string = "     listen first      "
print(space_string.rstrip())  # removes spaces from the right side
source_string = "interoperability"
print(source_string.rstrip("yt"))  # removes "y" or "t" or "yt" or "ty" from the right side
```

**split method**

string.split
Return a list of the words in the string, separated by the delimiter string. Default delimiter-separator is space.

```python
print("1,2,3".split())
print("1,2,3".split(','))  # indexliyoruz burada 0, 1 ve 2 olarak.
print("1,2,3".split(',', maxsplit= 1))
print("1,2,3".split('aa'))
```

```python
alfabe = "abcçdefgğhıijklmnoöprsştuüvyz"
print(alfabe.split())
print(alfabe.split("i"))  # i'yi cıkarıp ayırır.
print(alfabe.split(',', maxsplit= 1))
harf_listesi = list(alfabe)
print(harf_listesi)
```

join fonksiyonu da ("".join(variable)) listeyi belirtilen separator ("separator") ile tekrar string yapar.

```python
text = "Her gün, her hafta, bilgime yeni bilgiler katıyorum, bu beni çok motive ediyor."
text2 = text.split()  # boşlukları kullanarak liste yapar
print(text2)
text3 = " _".join(text2)  # listeyi geri string yapar. burada her birleştirdiği index arasına " _" ekler.
print(text3)
text4 = text3.split(",")  # virgül gördüğü yerlerden itibaren slice ederek liste yapar
print(text4)
print(",#".join(text4))  # her indexi aralarına ",#" koyarak tekrar string yapar.
```
