**Structure of the 'if' Statements**

In some cases, your program needs to execute some part of the code only if a specific condition is true.

The if statements check the condition. **The condition is always a Boolean expression**, that is, its value equals either True or False.

If it evaluates to True, it executes **the body of the if statement**. If it evaluates to False, it skips the body. The **condition ends with a colon and a new line starts with an indentation** (four spaces).


```python
empty_seat = 14
if empty_seat > 3:  # in this case, 14>3=True, so the body will execute
    print('there is still seat to sit')
```

**'if-else' Statements**

An if-else statement is another kind of conditional statements in Python. It is used with an additional keyword: else.

**else works like an if statement. If none of the conditions in if are ensured, "else" will be used to specify all remaining conditions**. 

else doesn't require any condition.

```python
number = 5
if number <= 3:    
    print("Number is smaller than or equal to 3") 
else:  # Optional clause (you can only have one else)
    print("Number is bigger than 3")
```

**'if-elif-else' Statements**

The **elif statement is used when it requires to specify several conditions in our program**.

In Python you can deﬁne a series of conditionals using :

**if for the ﬁrst one**,
**elif for the rest, up until the ﬁnal (optional)**,
**else for anything not caught by the other conditionals**

Condition1 is checked first, and if the result is True the body1 in the 'if' statement' will be executed. If not, condition2 in the 'elif' statement is checked. If the result is True, body2 will be executed, if not, body3 in the last control point 'else' will be executed.

```python
weight = 80
if weight > 100:
    print("That's too heavy!")
elif weight > 75:
    print("I can lift that!")
else:
    print("That's too light!")
```

 The keyword ‘elif’ is short for ‘else if’, and is useful to avoid excessive indentation. An if … elif … elif … sequence is a substitute for the switch or case statements found in other languages.

```python
audience = "baby"
if audience == "kid":
    print("it is free to go to cinema")
elif audience == "teen":
    print("discounted price!")
elif audience == "adult":
    print("normal price")
else:
    print("No such audience, stay at your home!")
```

3. Eğer başkan katıldıysa bir başkan yardımcısının katılımı yeterlidir. Başkan katılmadıysa 2 başkan yardımcısı da hazır olmalıdır. 

Aşağıdaki bilgileri kullanarak "toplantı yeter sayısına ulaşılmıştır" ya da "toplantı yeter sayısına ulaşılamamıştır" output'unu print ediniz.

Yönetim kurulu üyeleri: Ahmet (Başkan), Veli, Jim , John, Kim (Başkan Yard), Julia (Başkan Yard), Nelson

Toplantıya katılanlar : John, Julia, Nelson, Kim 


```python
executive_members = ['Ahmet', 'Veli', 'Jim', 'John', 'Kim', 'Julia', 'Nelson']
participants = ['John', 'Julia', 'Nelson', 'Kim']

if len(participants) >= 4 and (set(participants) - set(executive_members) == set()):
    if (executive_members [0] in participants and (executive_members [2] 
                                                   or executive_members [5]) in participants):
      print("toplantı yeter sayısına ulaşılmıştır")
    elif (executive_members [0] not in participants and 
          (executive_members[2] and executive_members[5]) in participants):
      print("toplantı yeter sayısına ulaşılmıştır")
        
else:
    print("Toplantı yeter sayısına ulaşılamamıştır!")
```

An integer is called as “Steven even” if the sum of its digits is even.
For example, 51 is a “Steven even” because the sum of 5 and 1 is even.
Write a program that takes an integer (between 10 and 99) from the user and
then prints whether the integer is:
Even and Steven even.
Not even but Steven even
Even but not Steven even
Neither even nor Steven even

```python

user_number = "24"
digits_sum = int(user_number[0]) + int(user_number[1])

if not user_number.isdigit() :
    print("Enter a valid number")
elif not (int(user_number) % 2) and digits_sum % 2 == 0 :  # bu su demek: (int(user_number) % 2 == 0) and digits_sum % 2 == 0 :
    print("Even and Steven even")
elif not (int(user_number) % 2) and digits_sum % 2 :  # int(user_number) % 2 == 0) and digits_sum % 2
    print("Even but not Steven even")
elif (int(user_number) % 2) and not digits_sum % 2:  # (int(user_number) % 2 == 1) and not digits_sum % 2:
    print("Not even but Steven even")
else : 
    print("Neither even nor Steven even")
```
